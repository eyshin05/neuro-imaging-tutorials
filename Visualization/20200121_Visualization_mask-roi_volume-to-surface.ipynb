```python
import nilearn.datasets
import nilearn.image
import nilearn.plotting
import nilearn.surface

import numpy as np
import matplotlib.pyplot as plt
```

```python
stat_map = nilearn.image.load_img('stat_map.nii.gz')
mask_img = nilearn.image.math_img(img1=stat_map, formula='img1 > 0')
nilearn.plotting.plot_glass_brain(mask_img)
```

```python
big_fsaverage = nilearn.datasets.fetch_surf_fsaverage('fsaverage')
```

```python
# sulc background data processing - to emphasize sulcus, add binarized data

bg_data = nilearn.surface.load_surf_data(big_fsaverage.sulc_right)
bg_data_bin = np.zeros(bg_data.shape)
bg_data_bin[bg_data > 0] = 1
bg_data_bin += bg_data
```

```python
texture = nilearn.surface.vol_to_surf(mask_img, big_fsaverage.pial_right, interpolation='nearest', radius=0.0)

fig = nilearn.plotting.plot_surf_roi(
    surf_mesh=big_fsaverage['infl_right'],
    roi_map=texture,
    hemi='right',
    view='posterior',
    cmap='Spectral',
    bg_map=bg_data_bin,
    darkness=0.7
)

# if you want to save this figure as a highres png file...
#fig.savefig('mask_right_posterior.png', dpi=300)
```

### In case of multiple ROIs

```python
rois = [nilearn.image.load_img('mot.nii.gz'),
        nilearn.image.load_img('som.nii.gz'),
        nilearn.image.load_img('sma.nii.gz')]
```

```python
# mot == 1, som == 2, sma == 3

all_roi_img = rois[0]

for i, img in enumerate(rois[1:]):
    all_roi_img = nilearn.image.math_img(img1=all_roi_img, img2=img, formula='img1 + img2*%d' % (i+2))
```

```python
texture = nilearn.surface.vol_to_surf(all_roi_img, big_fsaverage.pial_left, interpolation='nearest', radius=0.0)

fig = nilearn.plotting.plot_surf_roi(
    surf_mesh=big_fsaverage['infl_left'],
    roi_map=texture,
    hemi='left',
    view='lateral',
    cmap='Spectral',
    bg_map=bg_data_bin,
    darkness=0.7
)
```

```python
fig = nilearn.plotting.plot_surf_roi(
    surf_mesh=big_fsaverage['infl_left'],
    roi_map=texture,
    hemi='left',
    view='medial',
    cmap='Spectral',
    bg_map=bg_data_bin,
    darkness=0.7
)
```
